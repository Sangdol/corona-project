```python
date_max = df['date'].max()
print(date_max)
```

```python
today = df[(df.continent == 'Europe') & (df.date == date_max)]
today
```

```python
# too many missing total_tests 
columns = ['location', 'date', 'total_cases', 'new_cases', 'total_cases_per_million', 'new_cases_per_million', 'total_tests', 'population']
selected = today[columns]
selected
```

```python
selected.sort_values('total_cases_per_million', ascending=False)
```
